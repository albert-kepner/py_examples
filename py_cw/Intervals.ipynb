```python
def calculate(rectangles):
    rectangles.sort() ## rectangles sorted by lower x coord
    x_coords, x_rects = get_x_coord_grouping(rectangles)
    ## print(x_coords)
    ## print(x_rects)
    total_area = 0
    for index, x_val in enumerate(x_coords):
        if index == len(x_coords) - 1:
            continue;
        x_high = x_coords[index + 1]
        y_intervals = extract_y(x_rects[x_val])
        y_sum = sum_of_intervals(y_intervals)
        total_area = total_area + (x_high - x_val) * y_sum
    return total_area
    
def get_x_coord_grouping(rectangles):
    ## rectangles are already sorted by starting x value.
    xc = set()
    for x0, _, x1, _ in rectangles:
        xc.add(x0)
        xc.add(x1)
    x_coords = list(xc)
    x_coords.sort() ## We now have a sorted list of all distinct x values at rectangle borders
    x_rects = {} ## dictionary to hold rectangles in effect starting at each x_coord.
    rect_pos = 0
    current_rects = []
    for index, x_val in enumerate(x_coords):
        dels = []
        for ix, r in enumerate(current_rects):
            if r[2] == x_val:
                dels.insert(0,ix)
        for ix in dels: ## remove rectangles ending at this x_coord
            del current_rects[ix]
        while rect_pos < len(rectangles) and rectangles[rect_pos][0] == x_val: ## add rectangles starting at this x_coord
            current_rects.append(rectangles[rect_pos])
            rect_pos = rect_pos + 1
        ## Now we know which rectangles are in effect, starting at the corrent x_coord
        x_rects[x_val] = current_rects.copy()
    return (x_coords, x_rects)
            
def extract_y(rectangles):
    return [ (y0, y1) for x0, y0, x1, y1 in rectangles ]
        
def sum_of_intervals(intervals):
    intervals.sort()
    combined = []
    remaining = intervals
    if len(intervals) == 0:
        return 0
    elif len(intervals) == 1:
        a = intervals[0]
        return a[1] - a[0]
        
    while len(remaining) > 1:
        a, b = remaining[:2]
        remaining = remaining[2:]
        c = combine(a,b)
        if c:
            next = c
        else:
            combined.append(a)
            next = b
        if len(remaining) > 0:
            remaining.insert(0,next)
        else:
            combined.append(next)
    total = sum(map(lambda x : x[1] - x[0], combined))
    return total
        
def combine(a, b):
    """
    Combine the two intervals, if they overlap,
    return the combined interval, return None otherwise.
    """

    # Assume a[0] <= b[0] because intervals are sorted by start
    if a[1] >= b[0]:
        ## New interval from a[0] to max (a[1],b[1])
        return (a[0], max(a[1], b[1]) )
    else:
        return None      
```
